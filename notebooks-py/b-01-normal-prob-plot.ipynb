## Import the modules

```python
import src.load as load
from scipy.stats import probplot
import seaborn as sns
import matplotlib.pyplot as plt
```

## Plot a normal probability plot

```python
sns.set_theme()
fig, ax = plt.subplots()
res = probplot(x=plasma["Level"], plot=ax)
plt.show()
```
