# Central Limit Theorem: An Illustration

Date: 2021-01-04

A visualisation of the Central Limit Theorem in action.

The cental limit theorem states that

> ... if you have a population with mean $\mu$ and standard deviation $\sigma$ and take sufficiently large random samples from the population with replacement, then the distribution of the sample means will be approximately normally distributed.

We will use the continuous Uniform distribution $U(0,1)$ for this visualisation.

```python
import scipy.stats as stats
from util.clt import clt
```

## Plot the samples

```python
# declare the distribution
u = stats.uniform(loc=0, scale=1)
```

```python
clt(a_dist=u,
    n=30,
    N=30,
    bins=50,
    plot_normal=True)
```

```python
clt(a_dist=u,
    n=30,
    N=100,
    bins=50,
    plot_normal=True)
```

```python
clt(a_dist=u,
    n=30,
    N=300,
    bins=50,
    plot_normal=True)
```

```python
clt(a_dist=u,
    n=30,
    N=3000,
    bins=50,
    plot_normal=True)
```
