# Central Limit Theorem: An Illustration

Date: 2021-01-04

A visualisation of the Central Limit Theorem in action.

The cental limit theorem states that

> ... if you have a population with mean $\mu$ and standard deviation $\sigma$ and take sufficiently large random samples from the population with replacement, then the distribution of the sample means will be approximately normally distributed.

We will use the continuous Uniform distribution $U(0,1)$ for this visualisation.

`clt(a_dist, n, N, bins)`
: Generates `N` random samples of size `n` from the distribution `a_dist`.
Calulate's each samples mean, and plots the sampling distribution of the mean on a histogram.
: The parameter `bins` controls the number of bins used in the histogram.
Changing this parameter will make the histogram less accurate, but improves the plot's fidelity for discrete distributions.
: Overlay the approximation of the distribution: $N(\mu, \sigma^{2}/n)$.

```python
import scipy.stats as stats
from util.clt import clt
```

## Plot the samples

```python
# declare the distribution
u = stats.uniform(loc=0, scale=1)
```

```python
clt(a_dist=u,
    n=30,
    N=10000,
    bins=50)
```

```python
clt(a_dist=stats.poisson(mu=5),
    n=50,
    N=10000,
    bins=20)
```
