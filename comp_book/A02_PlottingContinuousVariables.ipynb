# A2. Plotting Continuous Variables

> In this chapter, you will learn how to obtain plots of continuous variables: frequency histograms in Subsection 2.1 and boxplots in Subsection 2.2.

We make use of the `query.py` script here.
(See **description**).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import util.query as query
```

```python
db = query.Database()  # creates a connection to the database
tbl = query.Table("Membership", db)  # returns the table as a list
df = tbl.toDF()  # returns the list as a DataFrame
```

For these three activities, we require the data to of type `float`.
Let us check the `dtypes` as they are now, and recast as required.

```python
df.dtypes
```

Each field has been imported as an `object`.
Let us recast the field as a `float`.

```python
df = df.astype(dtype={'Percentage': 'float'})
```

## A2.1 Frequency histograms

```python
fig = sns.histplot(data=df, x="Percentage")
fig.set(title="Adult Membership of Sports Club by Percentage")
plt.show()
```

We can alter the binwidth by using the `binwidth` parameter.
(Reference: **Add**)

```python
fig = sns.histplot(data=df, x="Percentage", binwidth=1)
fig.set(title="Adult Membership of Sports Club by Percentage")
plt.show()
```

## A2.2 Boxplots

```python
fig = sns.boxplot(data=df, x='Percentage')
fig.set(title="Adult Membership of Sports Club by Percentage")
plt.show()
```
