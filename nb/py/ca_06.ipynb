Using `sns.barplot()` to plot a bar chart of aggregated data.
We also sort the data to change the order of the `sns.barplot()`.

Ref: 

## Package import

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

## Data import

```python
df = pd.read_csv("../data/workforce.csv")

df.head()
```

## Plot the data

```python
f, ax = plt.subplots()

sns.barplot(data=df,
            x="Occupation type",
            y="Total",
            color="royalblue")

ax.set(title="Total workforce by Occupation type",
       xlabel="",
       ylabel="Frequency (millions)")

plt.xticks(rotation=70)  # rotate the x-axis labels
plt.tight_layout()       # pull the plot closer togethers
plt.show()
```

## Changing the order or the bars

```python
# order the DataFrame in descending order by total
ordered_df = df.sort_values("Total", ascending=False)

f, ax = plt.subplots()

sns.barplot(data=ordered_df,
            x="Occupation type",
            y="Total",
            color="royalblue")

ax.set(title="Total workforce by Occupation type",
       xlabel="",
       ylabel="Frequency (millions)")

plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```
