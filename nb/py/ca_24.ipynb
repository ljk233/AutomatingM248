Illustrate the *settling down* phenomenon using a discrete uniform distribution.

We should see that, as the sample size insreases, the plot becomes more like the discrete uniform.

```python
from scipy.stats import poisson, expon
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")  # set the theme for sns


def poisson_generate_and_plot_sample(mu: float, N: int) -> None:
    '''Generates a random sample from the Poisson(mu) distribution, and
    plots the distribution as a bar chart.'''

    # generate a random sample of size N from Poisson(mu)
    r = poisson.rvs(mu, size=N)

    f, ax = plt.subplots()

    sns.countplot(x=r,
                  color="royalblue",
                  ax=ax)

    ax.set(xlabel="X",
           ylabel="Frequency")
    
    plt.show()
```

First we plot the p.m.f. of **Poisson(10)**

```python
fig, ax = plt.subplots(1, 1)
# generate the range
x = np.arange(poisson.ppf(0.01, 10),
              poisson.ppf(0.99, 10))

sns.barplot(x=x,
            y=poisson.pmf(10, x),
            color="royalblue")

plt.show()
```

```python
poisson_generate_and_plot_sample(10, 10)
```

```python
poisson_generate_and_plot_sample(10, 100)
```

```python
poisson_generate_and_plot_sample(10, 300)
```

```python
poisson_generate_and_plot_sample(10, 1000)
```
