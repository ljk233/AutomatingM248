Using `sns.histplot()` to plot a frequency histogram of some data.

## Package import

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

## Data import

```python
# import the data
df = pd.read_csv("../data/membership.csv")

# preview data
df.head()
```

## Plot a default frequency histogram

```python
f, ax = plt.subplots()
sns.histplot(data=df,
             x="Percentage",
             edgecolor="none",  # remove the borders
             color="royalblue")

ax.set(title="Frequency histogram of Percentage",
       xlabel="Percentage",
       ylabel="Frequency")

plt.show()
```

## Plot a frequency histogram with modified binwidth

```python
f, ax = plt.subplots()
sns.histplot(data=df,
             x="Percentage",
             binwidth=1,        # set bins to width 1
             edgecolor="none",  # remove the borders
             color="royalblue")

ax.set(title="Frequency histogram of Percentage",
       xlabel="Percentage",
       ylabel="Frequency")

plt.show()
```
