# A7-1. Settling down and the Poisson distribution

-----


```python
import scipy.stats as stats
import seaborn as sns
import numpy as np
```

Let us generate and plot a series of random variables from the distribution $X \sim \text{Poisson}(8)$, so we can see how the distribution *settles down* as sample size grows.


We will use:

~~~python
# Calculate the pmf
rv = poisson(8)
rv.pmf(x)

# Generate N-size list of a rv
r = stats.poisson.rvs(mu=8, size=N)

# Plot the generated list
ax = sns.countplot(x=r)
~~~

## Plotting $p(x)$


```python
# generated array of x 0:30
X = np.arange(30)
# declate Poisson(8)
rv = stats.poisson(8)
p = list()

# for each x in X
for i in X:
    # calculate p(x), append to list
    p.append(rv.pmf(i))

# plot p(x)
ax1 = sns.barplot(x=X, y=p, palette="Blues_d")
ax1.set(xlabel='x', ylabel='p(x)', title='Poisson(8)')
```

It can be seen that $N$ increases, the distribution approaches the probability distribution for $\text{Poisson}(8)$.

## Generating $N = 10$ random variables

```python
r = stats.poisson.rvs(mu=8, size=10)
ax2 = sns.countplot(x=r, palette="Blues_d")
ax2.set(xlabel='x', ylabel='Frequency')
```
