Visualising the *settling down* phenomenon of $X \sim \text{Poisson}(8)$.

> *In a long sequence of repetitions of a study or experiment, bar charts and histograms of samples tend to settle down towards probability distributions*


## Plotting p(x)


```python
import scipy.stats as stats
import seaborn as sns
import numpy as np
```

```python
# generated array of x 0:30
X = np.arange(30)

# declare Poisson(8)
rv = stats.poisson(8)
p = list()

# for each x in X
for i in X:
    # calculate p(x), append to list
    p.append(rv.pmf(i))

# plot p(x)
ax1 = sns.barplot(x=X, y=p, palette="Blues_d")
ax1.set(xlabel='x', ylabel='p(x)', title='Poisson(8)')
```

## Generating and Plotting N random variables Poisson(8)

```python
# simulate N=10
r = stats.poisson.rvs(mu=8, size=10)
ax2 = sns.countplot(x=r, palette="Blues_d")
ax2.set(xlabel='x', ylabel='Frequency',
        title='Poisson(8), N=10')
```

```python
# simulate N=100
r = stats.poisson.rvs(mu=8, size=100)
ax3 = sns.countplot(x=r, palette="Blues_d")
ax3.set(xlabel='x', ylabel='Frequency',
        title='Poisson(8), N=100')
```

```python
# simulate N=1000
r = stats.poisson.rvs(mu=8, size=1000)
ax4 = sns.countplot(x=r, palette="Blues_d")
ax4.set(xlabel='x', ylabel='Frequency',
        title='Poisson(8), N=1000')
```
