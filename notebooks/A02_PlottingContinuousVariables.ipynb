```python
import pandas as pd
import seaborn as sns
import queryM248 as query
```

## 2. Data import

```python
# connect to the db
db = query.Database("sets")

# query the table
table = query.Table("Membership", db)

# declare the DataFrame
df1 = table.toDF()
```

## 3. Data wrangling

```python
# check the dtypes
df1.dtypes
```

The percentage columns has come through as an object, so we need to recast percentage from `object` $\to$ `float`

```python
df1 = df1.astype({'Percentage': 'float'})
```

```python
# check the dtypes
df1.dtypes
```

## 3. Plot the default histogram

```python
# Declare the defaults histogram
fig1 = sns.histplot(df1,
                   x="Percentage")

# Label ax1 and y-axis
fig1.set(ylabel="Frequency",
        title="Adult Membership of Sports Club by Percentage")
```

## 4. Plot another histogram with custom bin widths

```python
# Declare another histogram as ax2, set binwidth to 1
fig2 = sns.histplot(df1,
                   x="Percentage",
                   binwidth=1)

# Label fig and y-axis
fig2.set(ylabel="Frequency",
        title="Adult Membership of Sports Club by Percentage")
```

## 5. Plot a boxplot

```python
# declare the boxplot
fig3 = sns.boxplot(data=df1,
                  x='Percentage')

# Label fig and x-axis
fig3.set(xlabel="Percentage",
         title="Adult Membership of Sports Club by Percentage")
```
