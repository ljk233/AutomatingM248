```python
from scripts.data import Data
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
workforce = Data.load_workforce()
```

```python
workforce.head()
```

```python
f, ax = plt.subplots()

sns.barplot(
    data=workforce,
    x="Occupation type",
    y="Total",
    color="royalblue"
)

ax.set(
    title="Total workforce by Occupation type",
    xlabel="",
    ylabel="Workers (millions)"
)

# rotate labels in x-axis
plt.xticks(rotation=70)

# extends the width of the plot
plt.tight_layout()

plt.show()
```

The order of the bars can be changed by altering the ordering of source `DataFrame`.

```python
# plot the data
f, ax = plt.subplots()

# order the DataFrame in descending order by total
sns.barplot(
    data=workforce.sort_values("Total", ascending=False),
    x="Occupation type",
    y="Total",
    color="royalblue"
)

ax.set(
    title="Total workforce by Occupation type",
    xlabel="",
    ylabel="Workers (millions)"
)

# rotate labels in x-axis
plt.xticks(rotation=70)

# extends the width of the plot
plt.tight_layout()

plt.show()
```
