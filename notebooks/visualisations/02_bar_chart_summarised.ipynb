# Visualisation: Bar chart (aggregated data)

## Summary

The aim of the notebook is to plot a sample of aggregated data as a bar chart.
The data was plotted twice: the first time unsorted, the second time sorted by descending order.

## Introduction

This analysis is interested in displaying the total numbers of workers in a selection of occupation-types as a frequency bar chart.
The sample shows the number of workers in each of the 9 occupartion types, with figures showing the number for each gender, as well as the total numbers.
The data is provided by the Open University *(activity 6, Computer Book A, M248: Analysing data),* who do not provide a reference for their source.

## Method

The data was displayed as a frequency bar chart twice: Once with the columns unsorted, and again with the columns sorted by descending order.
The visualisation was done using `Python` using the `pandas`, `seaborn`, and `matplotlib` packages.

## Result

### Setup the notebook

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# set the seaborn theme
sns.set_theme()
```

```python
# import the data
df_workforce = pd.read_csv("..\\..\\data\\raw\\workforce.csv")
```

### Describe the data

```python
df_workforce.describe().T
```

### Visualise the data

#### Unsorted

```python
f, ax = plt.subplots()
sns.barplot(data=df_workforce, x="Occupation type", y="Total")
ax.set(xlabel="", ylabel="Workers (millions)")
# rotate labels in x-axis
plt.xticks(rotation=70)
# extends the width of the plot
plt.tight_layout()
plt.show()
```

#### Sorted by descending order

The order of the bars can be changed by altering the ordering of source `DataFrame`.

```python
# order the DataFrame in descending order by total
sorted_df = df_workforce.sort_values("Total", ascending=False)
```

```python
# plot the data
f, ax = plt.subplots()
sns.barplot(data=sorted_df, x="Occupation type", y="Total")
ax.set(xlabel="", ylabel="Workers (millions)")
# rotate labels in x-axis
plt.xticks(rotation=70)
# extends the width of the plot
plt.tight_layout()
plt.show()
```
