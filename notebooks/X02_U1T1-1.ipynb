```python
import M248.query as query
import seaborn as sns
```

```python
def addDescCol(df: object, col: str, a_dict: dict) -> object:
    '''
    Add a description column, replacing keys of original col
    with values in a_dict.
    '''

    # copy the column to a new column
    desc_col: str = 'desc_' + col
    df[desc_col] = df[col]

    # map the values in the descCol
    df[desc_col].replace(a_dict, inplace=True)

    return df
```

```python
# connect to db
db: object = query.Database("sets")
# query tables
bikeDaily = query.Table("BikeRentalDailyDC", db)
# output tables to df
df_bikeDaily = bikeDaily.toDF()
```

```python
# always check the dtypes
df_bikeDaily.dtypes
```

The import has brought through the data as objects, so we recast the data as appropriate.

Note, ordinal data was cast to integers (preserves ordering)

```python
# Recast dtypes
newTypes = {'instant': 'int',
            'dteday': 'datetime64[ns]',
            'season': 'int',
            'yr': 'int',
            'mnth': 'int',
            'holiday': 'category',
            'weekday': 'int',
            'workingday': 'category',
            'weathersit': 'category',
            'temp': 'float',
            'atemp': 'float',
            'hum': 'float',
            'windspeed': 'float',
            'casual': 'int',
            'registered': 'int',
            'dailycount': 'int'}
df_bikeDaily = df_bikeDaily.astype(newTypes)
```

The data includes some index columns, so add a description column

- `season`
- `yr`
- `mnth`
- `weekday`
- `workingday`

A utility function `addDescCol()` was written to support the mapping process


```python
# Add mapping columns
dictSeason = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df_bikeDaily = addDescCol(df_bikeDaily, 'season', dictSeason)

dictYr = {0: 2011, 1: 2012}
df_bikeDaily = addDescCol(df_bikeDaily, 'yr', dictYr)

dictMnth = {1: 'JAN', 2: 'FEB', 3: 'MAR', 4: 'APR',
            5: 'MAY', 6: 'JUN', 7: 'JUL', 8: 'AUG',
            9: 'SEP', 10: 'OCT', 11: 'NOV', 12: 'DEC'}
df_bikeDaily = addDescCol(df_bikeDaily, 'mnth', dictMnth)

dictWeekday = {0: 'SUN', 1: 'MON', 2: 'TUES', 3: 'WEDS',
               4: 'THURS', 5: 'FRI', 6: 'SAT'}
df_bikeDaily = addDescCol(df_bikeDaily, 'weekday', dictWeekday)

dictWorkingday = {'0': 'Non-working day', '1': 'Working day'}
df_bikeDaily = addDescCol(df_bikeDaily, 'workingday', dictWorkingday)
```

## Data visualisations

### Usage variation over the time

```python
fig1 = sns.lineplot(x=df_bikeDaily['dteday'],
                    y=df_bikeDaily['registered'])

fig1.set(xlabel='date',
         ylabel='Registered Users',
         title='Number of Registered Users by Date')
```

### Number of registered users by month

Notice that `SNS` has introduced shaded areas around the lines, representing the range of data.

```python
fig2 = sns.lineplot(x=df_bikeDaily['desc_mnth'],
                    y=df_bikeDaily['registered'],
                    hue=df_bikeDaily['desc_yr'])

fig2.set(xlabel='Month',
         ylabel='Registered Users',
         title='Number of Registered Users by Month')
```

### Distribution of users by weekday

```python
fig3 = sns.boxplot(x=df_bikeDaily['registered'],
                   y=df_bikeDaily['desc_weekday'])

fig3.set(xlabel='Registered Users',
         ylabel='Weekday',
         title='Distribution of Registered Users by Weekday')
```

### Distruction of users by type of day (working vs. non-working)

```python
fig4 = sns.boxplot(x=df_bikeDaily['registered'],
                   y=df_bikeDaily['desc_workingday'])

fig4.set(xlabel='Registered Users',
         ylabel='Type of Day',
         title='Distribution of Registered Users by Type of Day')
```

### Registered daily use against daily temperatures

```python
fig5 = sns.scatterplot(x=df_bikeDaily['temp'],
                       y=df_bikeDaily['registered'])

fig5.set(xlabel='Registered Users',
         ylabel='Frequency',
         title='Number of Registered Users against Temperature')
```
