# Central Limit Theorem: An Illustration

date: 2020-11-08

A visualisation of the Central Limit Theorem, showing that as the number of sample means increases, the distribution more closely resembles the Normal distribution.

We will use the continuous Uniform distribution $U(0,100)$ for this visualisation.

## Workflow

1. Declare the distribution
2. Generate a random sample from distribution
3. Calculate the mean of the sample
4. Add mean to the list
5. Plot histogram of the list

```python
import scipy.stats as stats
import M248.support as m248
```

```python
# declare the distribution
u = stats.uniform(loc=0, scale=100)
```
