## 1. Hypotheses

Let the hypotheses be

$$
\begin{aligned}
&H_{0} : \mu_{H} = \mu_{A} \\
&H_{1} : \mu_{H} > \mu_{A},
\end{aligned}
$$

where $\mu_{H}, \mu_{A}$ are the expected number of goals scored by home and away teams respectively.

## 2. Methodology

The data will be described using a table of summary results, followed by plot of the average number of goals by home and away teams.
We then perform a $z$-test to test the hypotheses.

To perform a $z$-test, we require the $z$-value, which is defined as

$$
Z = \frac{\mu - \mu_{0}}{S/\sqrt{n}}.
$$

In this study, we know that:

$\mu = \mu_{H} \simeq \overline{X}_{H} \simeq 1.57$;

$\mu_{0} \simeq \mu_{A} \simeq \overline{X}_{A} \simeq  1.25$;

$S = S_{A} \simeq 1.18$;

$n = 380$.

We can then restate the value for $Z$ as

$$
Z \simeq \frac{\overline{X}_{H} - \overline{X}_{A}}{S_{A}/\sqrt{n}}.
$$

*Note than in our calculations we used the exact values rather than the approximations.*

This will give us a $z$-value of approximately 5.22.

## 5. Discussion

With a $p$-value < 0.0000001, there is very strong evidence against the null hypothesis.
We therefore reject the null hypothesis and conclude that the mean number of goals scored by the home team is not equal to mean number of goals scored by the away teams.
On the basis of the sample, it seems probable that the expected number of goals by the home team is greater than that of the away team.
