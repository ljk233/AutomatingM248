## Introduction

The pioneering statistician W.S. Gosset (who published his work under the pseudonym 'Student') analysed data on sleep gains (in hours) of ten patients after they had taken each of the drugs L-hyoscyamine hydrobromide and D-hyoscyamine hydrobromide.

In this report we will use a $t$-test to test the null hypothesis that the underlying mean sleep gain for patients who have taken L-hyoscyamine hydrobromide is zero, against the one-sided alternative hypothesis that the mean sleep gain is greater than zero. (1)

## Hypotheses

Let $S$ be the sleep gain for patients who have taken L-hyoscyamine hydrobromide.
Then the hypotheses shall be

$$
\begin{aligned}
  &H_{0} : \mu_{S} = 0 \\
  &H_{0} : \mu_{S} > 0.
\end{aligned}
$$

## Results

The expected sleep gain for patients who were administered L-hyoscyamine hydrobromide was 2.33, with 95% $t$-interval (0.90, 3.76).
A one-sided $t$-test was performed, which yielded a test statistic $t \simeq 3.6799$ on degrees of freedom 9 with $p$-value 0.0025.

## Discussion

With $p \in (0.01, 0.05], we conclude that there is moderate against the null hypothesis.
We concude that the study provides that the sleep gained by patients who where administered L-hyoscyamine hydrobromide was greater than 0 (unit?).

```python
# declare sampling distribution of the mean
sampling_dist = t(
    df=sample["Ltreatment"].size-1,
    loc=sample["Ltreatment"].mean(),
    scale=sample["Ltreatment"].std()/(sample["Ltreatment"].size**0.5)
)

# get ci95 attendance
sampling_dist.interval(0.95)
```

```python
# normal probability plot
f, ax = plt.subplots()
probplot(
    sample["Ltreatment"],
    plot=ax
)
plt.show()
```

```python
# t-test
ttest_1samp(
    a=sample["Ltreatment"],
    popmean=0,
    alternative="greater"
)
```
